# 단순선형회귀

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,3,4,5,6])

# 기울기, 절편 구하기
beta_1 = np.sum((x-np.mean(x)) * (y-np.mean(y))) / np.sum((x-np.mean(x))**2)
beta_2 = np.mean(y) - beta_1 * np.mean(x)

print('기울기: ', beta_1)
print('절편: ', beta_2)
```

```python
plt.scatter(x,y)
plt.plot(x,beta_1 + beta_2 * x, color='red')
plt.show()
```
