```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
boston = load_boston()
x = boston.data
y = boston.target

scaler = StandardScaler()
x = scaler.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)

print(len(x_train), len(y_train), len(x_test), len(y_test))
```

```python
input_dim = x.shape[1]
print(input_dim)
output_dim = 1
learning_rate = 0.00000005
num_epochs = 3000

model = nn.Linear(input_dim,output_dim)
```

```python
optimizer = {'SGD' : optim.SGD(model.parameters(), lr=learning_rate),
            'Momentum' : optim.SGD(model.parameters(), lr=learning_rate, momentum=0.9),
            'Adagrad' : optim.Adagrad(model.parameters(), lr=learning_rate),
            'RMSprop' : optim.RMSprop(model.parameters(), lr=learning_rate),
            'Adam' : optim.Adam(model.parameters(), lr=learning_rate)}
```

```python
for optimizer_name, optimizer in optimizer.items():
    criterion = nn.MSELoss()
    optimizer.zero_grad()
        
    for epoch in range(num_epochs):
        inputs = torch.from_numpy(x_train).float()
        labels = torch.from_numpy(y_train).float()
            
        # Foward Pass
        outputs = model(inputs)
        loss = criterion(outputs, labels)
            
        # backward & optimize
        loss.backward()
        optimizer.step()
            
        if (epoch+1) % 100 == 0:
            print(f'{optimizer_name} - Epoch [{epoch+1}], Loss: {loss.item():.4f}')
            
```
