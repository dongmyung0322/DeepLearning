```python
import torch
import matplotlib.pyplot as plt
```

# 가중치

```python
# 2차원 가중치 행렬
w = torch.randn(3,3)

#1차원으로 변환
w_flatten = w.flatten()
```

```python
plt.hist(w_flatten, bins=20)
plt.xlabel('Weight value')
plt.ylabel('Frequency')
plt.show()
```

# 편향

```python
# 편향을 랜덤으로 초기화
b = torch.randn(1)

plt.hist(b.detach().numpy(), bins=20)   # .detach().numpy() 텐서를 넘파이 배열로 변환
plt.xlabel('Bias value')
plt.ylabel('Frequency')
plt.show()
```
