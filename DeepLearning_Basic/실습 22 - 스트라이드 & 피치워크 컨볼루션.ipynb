```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2
```

# 스트라이드 컨볼루션

```python
input_data = torch.randn(1,1,28,28)

# 콘볼루션 레이어 정의
conv = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, stride=2, padding=1)
output = conv(input_data)
print(output)
```

```python
plt.subplot(1,2,1)
plt.imshow(input_data.squeeze(), cmap='gray')
plt.title('input')
plt.subplot(1,2,2)
plt.imshow(output.squeeze().detach().numpy()[0], cmap='gray')
plt.title('output')
```

# 스트라이드 컨볼루션 On 이미지

```python
image_path = './data/cat.png'
#input_img = Image.open(image_path).convert('L')
input_img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
# 이미지의 텐서변환
input_data = torch.unsqueeze(torch.from_numpy(np.array(input_img)), dim=0).float()

conv = nn.Conv2d(in_channels=1, out_channels=1, kernel_size=3, stride=2)
output = conv(input_data)
```

```python
plt.subplot(1,2,1)
plt.imshow(input_data.squeeze().detach().numpy(), cmap='gray')
plt.title('input')
plt.subplot(1,2,2)
plt.imshow(output.squeeze().detach().numpy(), cmap='gray')
plt.title('output')

plt.tight_layout()
plt.show()
```

# 피치워크 컨볼루션

```python
input_data = torch.randn(1,1,32,32)

# 콘볼루션 레이어 정의
conv = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, dilation=2)
output = conv(input_data)
print(output)
```
