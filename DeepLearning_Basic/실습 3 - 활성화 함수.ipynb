```python
import numpy as np
import matplotlib.pyplot as plt
```

# Sigmoid

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-10,10,100)  # -10~10사이 랜덤값 100개
y = sigmoid(x)

plt.plot(x,y)
plt.title('sigmoide function')
plt.xlabel('input')
plt.ylabel('output')
plt.show()
```

# Hyperbolic Tangent

```python
def tahn(x):
    return np.tanh(x)
```

```python
x = np.linspace(-10,10,100)  
y = tahn(x)

plt.plot(x,y)
plt.title('hyperbolic tangent function')
plt.xlabel('input')
plt.ylabel('output')
plt.show()
```

# ReLU

```python
def relu(x):
    return np.maximum(0,x)
```

```python
x = np.linspace(-10,10,100)  
y = relu(x)

plt.plot(x,y)
plt.title('relu function')
plt.xlabel('input')
plt.ylabel('output')
plt.show()
```

# Leaky ReLU

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x>=0, x, x*alpha)

x = np.linspace(-10,10,100)  
y = leaky_relu(x)

plt.plot(x,y)
plt.title('leaky relu function')
plt.xlabel('input')
plt.ylabel('output')
plt.show()
```

# Softmax

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))
```

```python
x = np.linspace(-10,10,100)  
y = softmax(x)

plt.plot(x,y)
plt.title('softmax function')
plt.xlabel('input')
plt.ylabel('output')
plt.show()
```
